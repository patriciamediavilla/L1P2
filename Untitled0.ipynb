```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#ALCOHOLç
X1 = datosok["alcohol"].values.reshape(-1, 1)
y1 = datos["quality"].values

# Dividimos en test y train
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=0)

regressor1 = LinearRegression()
regressor1.fit(X_train1, y_train1)

# Predecir los datos de prueba
y_pred1 = regressor1.predict(X_test1)

# Evaluar el rendimiento del modelo
R1 = regressor1.score(X_test1, y_test1)
mae1 = metrics.mean_absolute_error(y_test1, y_pred1)
mse1 = metrics.mean_squared_error(y_test1, y_pred1)


#VOLATILE ACIDITY
X2 = datosok["volatile acidity"].values.reshape(-1, 1)
y2 = datos["quality"].values

# Dividimos en test y train
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=0)

regressor2 = LinearRegression()
regressor2.fit(X_train2, y_train2)

# Predecir los datos de prueba
y_pred2 = regressor2.predict(X_test2)

# Evaluar el rendimiento del modelo
R2 = regressor2.score(X_test2, y_test2)
mae2 = metrics.mean_absolute_error(y_test2, y_pred2)
mse2 = metrics.mean_squared_error(y_test2, y_pred2)


#SULPHATES
X3 = datosok["sulphates"].values.reshape(-1, 1)
y3 = datos["quality"].values

# Dividimos en test y train
X_train3, X_test3, y_train3, y_test3 = train_test_split(X3, y3, test_size=0.2, random_state=0)

regressor3 = LinearRegression()
regressor3.fit(X_train3, y_train3)

# Predecir los datos de prueba
y_pred3 = regressor3.predict(X_test3)

# Evaluar el rendimiento del modelo
R3 = regressor3.score(X_test3, y_test3)
mae3 = metrics.mean_absolute_error(y_test3, y_pred3)
mse3 = metrics.mean_squared_error(y_test3, y_pred3)


#CITRIC ACID
X4 = datosok["citric acid"].values.reshape(-1, 1)
y4 = datos["quality"].values

# Dividimos en test y train
X_train4, X_test4, y_train4, y_test4 = train_test_split(X4, y4, test_size=0.2, random_state=0)

regressor4 = LinearRegression()
regressor4.fit(X_train4, y_train4)

# Predecir los datos de prueba
y_pred4 = regressor4.predict(X_test4)

# Evaluar el rendimiento del modelo
R4 = regressor4.score(X_test4, y_test4)
mae4 = metrics.mean_absolute_error(y_test4, y_pred4)
mse4 = metrics.mean_squared_error(y_test4, y_pred4)


#TOTAL SULFUR DIOXIDE
X5 = datosok["total sulfur dioxide"].values.reshape(-1, 1)
y5 = datos["quality"].values

# Dividimos en test y train
X_train5, X_test5, y_train5, y_test5 = train_test_split(X5, y5, test_size=0.2, random_state=0)

regressor5 = LinearRegression()
regressor5.fit(X_train5, y_train5)

# Predecir los datos de prueba
y_pred5 = regressor5.predict(X_test5)

# Evaluar el rendimiento del modelo
R5 = regressor5.score(X_test5, y_test5)
mae5 = metrics.mean_absolute_error(y_test5, y_pred5)
mse5 = metrics.mean_squared_error(y_test5, y_pred5)
```

ARBOLES DE DECISION

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

bins = [0, 2, 4, 8, 10] 
labels = ['muy malo', 'malo', 'bueno', 'muy bueno'] 
datos['bins'] = pd.cut(datos['quality'], bins=bins, labels=labels)

X = datos[['alcohol', 'volatile acidity', 'sulphates']]
y = datos['bins']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

clf = DecisionTreeClassifier(max_depth=4, random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

# Calcular la exactitud (accuracy) del modelo
acc = accuracy_score(y_test, y_pred)
print(f"Exactitud del modelo: {acc:.3f}")
```
